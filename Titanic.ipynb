```python
df = pd.read_csv('train_and_test2.xls', low_memory=False)
display(df.columns)
display(df.head())
```

```python
display(df)
df = df.rename(columns={'2urvived': 'Survived', 'sibsp': 'SibSp'})
display(df.columns)
```

```python
xs = df.drop('Survived', axis=1)
y = df['Survived']
```

```python
from sklearn.ensemble import RandomForestClassifier

def rf(xs, y, n_estimators=40, max_samples=1000,
       max_features=0.5, min_samples_leaf=5):
    return RandomForestClassifier(
        n_jobs=-1,
        n_estimators=n_estimators,
        max_samples=max_samples,
        max_features=max_features,
        min_samples_leaf=min_samples_leaf,
        oob_score=True
    ).fit(xs, y)
```

```python
trn_idx, val_idx = RandomSplitter(valid_pct=0.2)(df)
train_xs, train_y = xs.iloc[trn_idx], y.iloc[trn_idx]
valid_xs, valid_y = xs.iloc[val_idx], y.iloc[val_idx]

m = rf(train_xs, train_y)

preds = np.stack([t.predict(valid_xs) for t in m.estimators_])

print(preds.shape)
preds.mean(0)
display(preds)
```

```python
from sklearn.metrics import accuracy_score

val_preds = m.predict(valid_xs)

acc = accuracy_score(valid_y, val_preds)
print(f"Validation Accuracy: {acc:.2%}")
```
